```python
#서울특별시 일반음식점 인허가 정보

import urllib.request 
import json 
import pandas as pd 
from pandas.io.json import json_normalize
```

```python
a=1
b=1000

df_mine = pd.DataFrame(columns=['APVPERMYMD','TRDSTATEGBN','TRDSTATENM','DCBYMD','SITEAREA','RDNWHLADDR','BPLCNM','UPTAENM','X','Y'])

while b < 460000 :
    url='http://openapi.seoul.go.kr:8088/5050435353726e643637446f4f5774/json/LOCALDATA_072404/'+str(a)+'/'+str(b)
    response = urllib.request.urlopen(url)
    json_str = response.read().decode("utf-8")
    json_object = json.loads(json_str)
    df = pd.json_normalize(json_object['LOCALDATA_072404']['row'])
    
    df_sel = df[['APVPERMYMD','TRDSTATEGBN','TRDSTATENM','DCBYMD','SITEAREA','RDNWHLADDR','BPLCNM','UPTAENM','X','Y']]
    
    df_mine = pd.concat([df_sel,df_mine])
    
    
    a = a + 1000
    b = b + 1000 
    
df
```

```python
df_mine.to_csv('./data/data1.csv')
```
