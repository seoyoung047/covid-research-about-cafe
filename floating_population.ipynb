```python
import pandas as pd
```

```python
b2017 = pd.read_csv('./data/2017b.csv')
b2017.info()
```

```python
f = b2017.fillna(method='ffill').iloc[:,:4]
```

```python
a = b2017.iloc[:,4:]
```

```python
fl2017 = pd.concat([f,a],axis=1)
```

```python
fl2017 = fl2017.loc[fl2017['구분'] == '계']
```

```python
fl2017 = fl2017.reset_index(drop=True)
```

```python
fl2017.iloc[:20,:]
```

```python
fl2017 = fl2017.drop(['구분'],axis=1)
```

```python
h_del = fl2017.loc[fl2017['행정동'] == '합계'].index
```

```python
fl2017 = fl2017.drop(h_del)
```

```python
s_del = fl2017.loc[fl2017['행정동'] == '소계'].index
```

```python
fl2017 = fl2017.drop(s_del)
```

```python
fl2017 = fl2017.reset_index(drop=True)
```

```python
fl2017['행정동'].unique()
```
