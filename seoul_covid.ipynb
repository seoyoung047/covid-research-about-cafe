```python
k=0
confirmed = []
for i in range(len(df)):
    for j in range(26):
        k = df.iloc[i,j+1] + k
        j += 1
    confirmed.append(k)
    k=0
    i += 1

len(confirmed)
```

```python
df2 = pd.DataFrame({'Date':df['자치구 기준일'] , 'confirmed':confirmed})
df2
```

```python
seoul_conf = pd.DataFrame({'Year':[],'Month':[],'confirmed':[],'confirmedadd':[]})
```

```python
y = 2020

c = 0
x = 0

ylist=[]
mlist=[]
xlist=[]
clist=[]

while y < 2022 :      #2020.2~2021.9

    for j in range(2,13) :
        m = j 
        df3 = df2[df2['Date'].dt.year == 2020]
        df3 = df3[df3['Date'].dt.month == m].reset_index(drop=True)

        x = df3.loc[0,'confirmed'] - c             # x = 매월 신규 확진자, c = 누적 확진자
        c = df3.loc[0,'confirmed']
        print(x,c)
        ylist.append(y)
        mlist.append(m)
        xlist.append(x)
        clist.append(c)
    y+=1
    
    for k in range(1,10) :
        m = k
        df3 = df2[df2['Date'].dt.year == 2021]
        df3 = df3[df3['Date'].dt.month == m].reset_index(drop=True)

        x = df3.loc[0,'confirmed'] - c             # x = 매월 신규 확진자, c = 누적 확진자
        c = df3.loc[0,'confirmed']
        print(x,c)
        ylist.append(y)
        mlist.append(m)
        xlist.append(x)
        clist.append(c)
    y+=1

        
seoul_conf['Year'] = ylist
seoul_conf['Month'] = mlist
seoul_conf['confirmed'] = xlist
seoul_conf['confirmedadd'] = clist
```

```python
seoul_conf['confirmed'] = seoul_conf['confirmed'].astype(int)
```

```python
seoul_conf['confirmedadd'] = seoul_conf['confirmedadd'].astype(int)
```
