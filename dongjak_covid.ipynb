```python
y = 2020

c = 0
x = 0

ylist=[]
mlist=[]
xlist=[]
clist=[]

while y < 2022 :      #2020.2~2021.9

    for j in range(2,13) :
        m = j 
        dj = dongjak_covid[dongjak_covid['Date'].dt.year == 2020]
        dj = dj[dj['Date'].dt.month == m].reset_index(drop=True)

        x = dj.loc[0,'Dongjak'] - c             # x = 매월 신규 확진자, c = 누적 확진자
        c = dj.loc[0,'Dongjak']
        print(x,c)
        ylist.append(y)
        mlist.append(m)
        xlist.append(x)
        clist.append(c)
    y+=1
    
    for k in range(1,10) :
        m = k
        dj = dongjak_covid[dongjak_covid['Date'].dt.year == 2021]
        dj = dj[dj['Date'].dt.month == m].reset_index(drop=True)

        x = dj.loc[0,'Dongjak'] - c             # x = 매월 신규 확진자, c = 누적 확진자
        c = dj.loc[0,'Dongjak']
        print(x,c)
        ylist.append(y)
        mlist.append(m)
        xlist.append(x)
        clist.append(c)
    y+=1

        
dj_covidm['Year'] = ylist
dj_covidm['Month'] = mlist
dj_covidm['confirmed'] = xlist
dj_covidm['confirmedadd'] = clist
```
